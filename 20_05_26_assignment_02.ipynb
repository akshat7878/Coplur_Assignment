```python
#Question 01

print("--Dictionary--")
myDict = {}
key = ['a', 'b', 'c', 'd', 'e']
val = [0, 1, 2, 3, 4]

#appending
for i in range(5):
    myDict[key[i]] = val[i]

print("key : val")
for key, val in myDict.items():
    print(f" {key}  :  {val}")

print()
print(myDict.keys())
print(myDict.values())
print()

myDict.update({'b': 3})
print(f"Updated Dictionary: {myDict}")

print(f"Popped item: {myDict.popitem()}")


print(f"\n{'='*60}\n")


print("--Tuple--")
tup = ()
print(tup, id(tup))

tup = (3,3)
print(tup, id(tup))

print(f"Occurence of 3: {tup.count(3)}")
print(f"Index of 3: {tup.index(3)}")


print(f"\n{'='*60}\n")


print("--Set--")
L = [1, 1, 3, 3, 4, 6, 6, 6]
st = set(L)
print(f"List: {L}\nSet: {st}")

st.add(7)
print(st)
```

```python
#Question 02
def add(n1, n2):
    return n1 + n2

def sub(n1, n2):
    return n1 - n2

def mul(n1, n2):
    return n1 * n2

def div(n1, n2):
    if n2 == 0:
        return ZeroDivisionError
    return n1 / n2

n1 = int(input("Enter first Number: "))
n2 = int(input("Enter second Number: "))

print(f"Addition: {add(n1, n2)}")
print(f"Subtraction: {sub(n1, n2)}")
print(f"Multiplication: {mul(n1, n2)}")
print(f"Division: {div(n1, n2)}")
```

```python
#Question 03

num = input("Enter the Number for Panlindrome Check: ")

temp = num[-1::-1]

if temp == num:
    print("It is a Panlindrome!")
else:
    print("It is Not a Panlindrome!")
```
