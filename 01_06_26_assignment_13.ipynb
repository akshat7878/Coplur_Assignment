1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 

```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])
arr
```

```python
arr = np.nan_to_num(arr, nan=0)
arr[[0,1], :] = arr[[1,0], :]
arr[:, [0,1,2,3]] = arr[:, [1,0,3,2]]

arr
```

2) Move axes of 3D array to new positions

3) Replace NaN values with average of columns

```python
arr = np.array([[1, 2, np.nan], [4, np.nan, 6], [7, 8, 9]])
print(arr)
print()

col_mean = np.nanmean(arr, axis=0)
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print(arr)
```

4) Replace negative value with zero in numpy array using replace 

```python
arr = np.array([5, -2, 8, -7, 10])
print(arr)
print()

arr[arr < 0] = 0
print(arr)
```

5) Study the following program import numpy as np # create a numpy 1d-arrays arr1 = np.array([3, 4]) arr2 = np.array([1, 0]) # find average of NumPy arrays avg = (arr1 + arr2) / 2 print("Average of NumPy arrays:\n", avg) -> Calculate average mean median mode values of two NumPy 2d-arrays 

```python
arr1 = np.array([3, 4])
arr2 = np.array([1, 0])

avg = (arr1 + arr2) / 2
print("Average of NumPy arrays:\n", avg)
```

```python
from scipy import stats
res = np.concatenate((arr1, arr2))
print("res Array:", res)
print()

print("\nMean =", np.mean(res))
print("Median =", np.median(res))
print("Mode =", stats.mode(res, axis=None))
```

6) Solve the following equation using linalg() and inverse Matrix method x - 2y + 3z = 9 -x + 3y - z = -6 2x - 5y + 5z = 17 

```python
arr1 = np.array([[1, -2, 3], [-1, 3, -1], [2, -5, 5]])
arr2 = np.array([9, -6, 17])

print(np.linalg.solve(arr1, arr2))
print()
inv_arr1 = np.linalg.inv(arr1)
print(np.dot(inv_arr1, arr2))
```
