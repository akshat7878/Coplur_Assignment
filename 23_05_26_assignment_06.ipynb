```python
#Question 01
import requests

def weather_details(city, api_key):
    url = "https://api.openweathermap.org/data/2.5/weather?"
    final_url = f"{url}q={city }&appid={api_key}&units=metric"

    try:
        response = requests.get(final_url)
        response.raise_for_status()
        data = response.json()

        temperature = data["main"]["temp"]
        humidity = data["main"]["humidity"]
        feels_like = data["main"]["feels_like"]
        visibility = data["visibility"]
        wind_speed = data["wind"]["speed"]
        Description = data["weather"][0]["description"]

        print(f"Weather in {city}:")
        print(f"Temperature: {temperature}°C (Feels like: {feels_like}°C)")
        print(f"Humidity: {humidity}%")
        print(f"visibility: {visibility}km")
        print(f"wind_speed: {wind_speed}km/h")
        print(f"Description: {Description}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching weather data: {e}")

api_key = "b8aeb299f93d443bd5cbeecdb72fc733"
city = input("Enter city name: ")

weather_details(city, api_key)
```

```python
#Question 02
import random

l = ["rock", "paper", "scissor"]

while True:

    c = int(input('''
Game Start......
1 YES
2 NO | Exit
'''))

    if c == 1:

        uscore = 0
        cscore = 0

        for a in range(3):

            userinput = int(input('''
1 ROCK
2 PAPER
3 SCISSOR
'''))

            if userinput == 1:
                uch = "rock"

            elif userinput == 2:
                uch = "paper"

            elif userinput == 3:
                uch = "scissor"

            else:
                print("Invalid Choice")
                continue

            cch = random.choice(l)

            print("Your Choice:", uch)
            print("Computer Choice:", cch)

            if uch == cch:
                print("Game Draw")
                uscore += 1
                cscore += 1

            elif ((uch == "rock" and cch == "scissor") or
                  (uch == "paper" and cch == "rock") or
                  (uch == "scissor" and cch == "paper")):

                uscore += 2
                print("You Win")

            else:
                cscore += 2
                print("You Lose")

        print("\nFinal Score")
        print("Your Score:", uscore)
        print("Computer Score:", cscore)
        print()

        if uscore == cscore:
            print("Game Drawn")

        elif uscore > cscore:
            print("You Won the Game")

        else:
            print("You Lose the Game")

    else:
        print("Game Exited")
        break
```

```python
#Question 03
import requests

def fetch_random_user_from_api():
    url = "https://api.freeapi.app/api/v1/public/randomusers/user/random"
    response = requests.get(url) # string
    data = response.json()

    if data["success"] and "data" in data:
        user_data = data["data"]

        username = user_data["login"]["username"]
        country = user_data["location"]["country"]

        return username, country
    else: 
        raise Exception("Failed to fetch user data")
    
for i in range(3):
    try:
        username, country = fetch_random_user_from_api()
        print(f"Username: {username} \nCountry: {country}\n")

    except Exception as e:
        print(str(e))
```
