```python
# 1) Repeat a tuple three times using * operator
t = (1, 2, 3)
print(t * 3)
```

```python
# 2) Join three tuples into one using + operator
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

new_tuple = t1 + t2 + t3
print(new_tuple)
```

```python
# 3) Check whether an element exists in a tuple
t = (10, 20, 30, 40)
element = 30

if element in t:
    print("Element exists")
else:
    print("Element does not exist")
```

```python
# 4) Find total, highest, and lowest without sum(), max(), min()
t = (12, 5, 18, 7, 25)

total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total:", total)
print("Highest:", highest)
print("Lowest:", lowest)
```

```python
# 5) Filter tuple values greater than 10
n = (3, 14, 7, 22, 9, 41, 18, 5)
filtered = tuple(x for x in n if x > 10)

print(filtered)
```

```python
# 6) Count elements in a set without len()
s = {"cat", "dog", "bird", "fish"}

count = 0
for item in s:
    count += 1

print("Number of elements:", count)
```

```python
# 7) Combine two sets into one containing unique elements
s1 = {1, 2, 3}
s2 = {3, 4, 5}

combined = s1 | s2
print(combined)
```

```python
# 8) Find common elements in both sets
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

common = s1 & s2
print(common)
```

```python
# 9) Find elements in either set but not both (symmetric difference)
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1 ^ s2
print(result)
```
