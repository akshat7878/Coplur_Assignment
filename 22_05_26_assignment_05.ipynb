```python
import csv

data = [
    ['name', 'address', 'mobile', 'email'],
    ['akshat', 'jagatpura', '4214211512', 'akshat@email.com'],
    ['vidhi', 'jagatpura', '3413413414', 'vidhi@email.com'],
    ['updesh', 'durgapura', '1421424212', 'updesh@email.com']
]

with open('Address Book.csv', 'w') as f:
    write = csv.writer(f)

    write.writerows(data)
```

```python
with open('Address Book.csv', 'r') as f:
    read = csv.reader(f)

    for row in read:
        print(row)
```

```python
#Question 02
import sqlite3

con = sqlite3.connect("Adresses.db")

sql = """
    CREATE TABLE AddressBook(
        Sr_No INTEGER PRIMARY KEY AUTOINCREMENT,
        Name VARCHAR(20), 
        Address VARCHAR(20), 
        Mobile VARCHAR(20), 
        Email VARCHAR(20)
    )
"""

con.execute(sql)
```

```python
sql = """
    INSERT INTO AddressBook('Name', 'Address', 'Mobile', 'Email') VALUES
    ('akshat', 'jagatpura', '4214211512', 'akshat@email.com'),
    ('vidhi', 'jagatpura', '3413413414', 'vidhi@email.com'),
    ('updesh', 'durgapura', '1421424212', 'updesh@email.com')
"""

con.execute(sql)
con.commit()
```

```python
sql = "SELECT* FROM AddressBook"
res = con.execute(sql)

for row in res:
    print(row)
```

```python
sql = "SELECT* FROM AddressBook WHERE Name= 'vidhi'"
res = con.execute(sql)

for row in res:
    print(row)
```
