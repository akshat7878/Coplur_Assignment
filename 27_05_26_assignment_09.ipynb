1) Explore more regex patterns Eg. The regex pattern used to validate email addresses, mobile no, string, and more

```python
import re

email = "abc@gmail.com"

pattern = r'.+@.+\..+'

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
mobile = "9876543210"

pattern = r'^[0-9]{10}$'

if re.match(pattern, mobile):
    print("Valid Number")
else:
    print("Invalid Number")
```

```python
name = "Akshat"

pattern = r'^[A-Za-z]+$'

if re.match(pattern, name):
    print("Valid String")
else:
    print("Invalid String")
```

```python
password = "abc123"

pattern = r'^[A-Za-z0-9]+$'

if re.match(pattern, password):
    print("Valid Password")
else:
    print("Invalid Password")
```

2) Explore more datetime function and uses in pandas

```python
import pandas as pd

dates = pd.to_datetime([
    '2025-01-01',
    '2025-02-15',
    '2025-03-20'
])

dates
```

```python
pd.Timestamp.now()
```

```python
df = pd.DataFrame({
    'Date': pd.to_datetime(['2025-01-01', '2025-06-15'])
})

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

df['Day_Name'] = df['Date'].dt.day_name()

df
```

```python
start = pd.to_datetime('2025-01-01')
end = pd.to_datetime('2025-01-10')

difference = end - start
difference
```
