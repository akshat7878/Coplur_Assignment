```python
#Question 01
def find_max1(a, b, c):
    return max(a, b, c)

def find_max2(a, b, c):
    return a if a > b and a > c else b if b > c else c

print(find_max1(4,5,3))
print(find_max2(4,5,3))
```

```python
#Question 02
def manipulate(*arr):
    newarr = []

    for i in range(len(arr)):
        if arr.count(arr[i]) == 1:
            newarr.append(arr[i])
    
    return newarr

print(manipulate(1,2,3,2,4,1))
```

```python
#Question 03
def multiply_list(arr):
    product = 1
    for i in arr:
        product *= i

    return product

l = [1,2,3,4]
print(multiply_list(l))
```

```python
#Question 04
def factorial1(num):
    if num == 1: return 1
    return num * factorial1(num-1)

def factorial2(num):
    fac = 1
    for i in range(1, num+1):
        fac *= i

    return fac

print(factorial1(7))
print(factorial2(6))

```

```python
#Question 05
def reverse(s):
    s = s[-1::-1]
    return s

s = 'hello there how are you'
print(reverse(s))
```

```python
#Question 06
def valid_num(num):
    if 0 <= num <= 100:
        return "valid"
    else:
        return "invalid"
    
print(valid_num(10))
print(valid_num(-10))
```

```python
#Question 07
def even_number(arr):
    newarr = []
    for i in arr:
        if not(i & 1):
            newarr.append(i)

    return newarr

l = [1,2,3,4,5,6,7,8]
print(even_number(l))
```

```python
#Question 08
import math
def prime(num):
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return f"{num} not a Prime Number"
        else:
            return f"{num} is a Prime Number"

print(prime(5))
print(prime(6))
```

```python
#Question 09
def count_letter(s):
    capital = small = 0
    for i in s:
        if 65 <= ord(i) <= 90:
            capital += 1
        elif 97 <= ord(i) <= 122:
            small += 1
    
    return capital, small

s = "Hello Everyone Today is Function's Day"
c, s = count_letter(s)
print(f"Capital Letter: {c}\nSmall Letter: {s}")
```
